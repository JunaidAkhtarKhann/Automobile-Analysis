# 1) Write a Python program to perform the following tasks :
a. Apply OneHot coding on Country column.
B. Apply Label encoding on purchased column (Data.csv have two categorical column the country column, and the purchased column).


```python
# Istall scikit learn
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
a4.columns
```

```python
# --- Part A: Apply OneHot Encoding on the 'Country' column ---

# Create a OneHotEncoder instance
onehotencoder = OneHotEncoder()

# Perform OneHot Encoding on the 'Country' column
# The result is a NumPy array, so convert it back to a DataFrame
encoded_country = pd.DataFrame(onehotencoder.fit_transform(a4[['Country']]).toarray(),
                               columns=onehotencoder.get_feature_names_out(['Country']))

# Concatenate the OneHotEncoded columns with the original DataFrame
df = pd.concat([a4, encoded_country], axis=1)


# Drop the original 'Country' column as it's encoded now
a4.drop('Country', axis=1, inplace=True)
```

```python
# --- Part B: Apply Label Encoding on the 'Purchased' column ---

# Create a LabelEncoder instance
labelencoder = LabelEncoder()

# Apply Label Encoding on the 'Purchased' column
a4['Department'] = labelencoder.fit_transform(a4['Department'])
```

```python
# Display the modified DataFrame
print("\nDataFrame after encoding:")
print(a4)
```

#Write a Python NumPy program to compute the weighted average along the specified axis of a given flattened array.

```python
import numpy as np

def weighted_average(array, weights):
    """
    Compute the weighted average of a 1D array using the provided weights.

    Parameters:
    array (numpy.ndarray): The flattened 1D array of values.
    weights (numpy.ndarray): The weights for each value in the array.

    Returns:
    float: The weighted average of the array.
    """
    # Ensure the array and weights have the same shape
    if array.shape != weights.shape:
        raise ValueError("Array and weights must have the same shape.")

    # Compute the weighted average
    weighted_sum = np.sum(array * weights)
    sum_of_weights = np.sum(weights)

    if sum_of_weights == 0:
        raise ValueError("Sum of weights must not be zero.")

    return weighted_sum / sum_of_weights

# Example usage
values = np.array([10, 20, 30, 40, 50])
weights = np.array([1, 2, 3, 4, 5])

# Compute weighted average
result = weighted_average(values, weights)
print("Weighted Average:", result)
```
